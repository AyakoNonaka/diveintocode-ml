## 【課題】100日後の米粒の数

100日後には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。
また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。
（縦軸が米の数、横軸が日数）

```python
# 初日に貰える米の数
n_grains = 1
# その日貰える米粒の数
total_grains = n_grains
# その日貰える米粒の数のリスト
list_n_grains =[n_grains]
# 累計
holded_n_grains = n_grains
# 貰える米粒の累計数のリスト
list_total_grains =[total_grains]

for _ in range(100):
    holded_n_grains = holded_n_grains  * 2
    list_n_grains.append(holded_n_grains)     
    total_grains = total_grains + holded_n_grains
    list_total_grains.append(total_grains)

print("100日後には合計で米粒をいくつもらっているのか:",total_grains/100000000,"億粒")    
    
# print("100日後には合計で米粒をいくつもらっているのか:".format(total_grains/1))

# print("厚さ： {:.2f}万キロメートル".format(folded_thickness/10000000))
```

## 【課題2】100日後以外の米粒の数

この逸話は30日だったり81日だったりと日数が曖昧です。
どの話にも対応できるように日数を引数とした関数を作成しましょう。
関数を作成した上で、機械学習コースを受講する期間ではいくつ米粒をもらえるかを答えてください。
前の問題同様に折れ線グラフも出力してください。

以下の雛形には関数としての枠組みだけではなく、引数や返り値についてのコメントも記載してあります。
参考にして、自分なりにも分かりやすくコメントを書き換えてください。

```python
def compute_sorori_shinzaemon(day=120):

    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """
    
    n_grains = 1
    total_grains = n_grains
    list_n_grains =[n_grains]
    holded_n_grains = n_grains
    list_total_grains =[total_grains]

# ある日に貰う米粒と、累計を計算
    
    for _ in range(day -1):
        holded_n_grains = holded_n_grains  * 2
        list_n_grains.append(holded_n_grains) 
        total_grains = total_grains + holded_n_grains
        list_total_grains.append(total_grains)

    return total_grains, list_n_grains, list_total_grains
```

```python
# 4ヶ月１２０日で貰える米粒を計算
total_grains, list_n_grains, list_total_grains = compute_sorori_shinzaemon(day=120)
print(len(list_total_grains),"日で貰える米粒は")
print(total_grains,"粒")
```

```python
# グラフを表示
plt.title("total_grains")
plt.xlabel("day")
plt.ylabel("total_grants")
plt.plot(list_total_grains) 
plt.show()
```
