## 【課題4】配列、勾配、グラフを作ろう

```python
# 関数を作る
x = npdata
y = x**2

# 座標の配列を作る
array_xy_task4 = [x, y]
array_xy_coordinate_task4 = np.array(array_xy_task4).T
array_xy_coordinate_task4
```

## 【課題6】最小値を求めよう

最後に具体例を計算してもらって締めくくりとしましょう！
y=x^2+5xの最小値を勾配降下法ではなく、微分法を使って求めてください。

【回答】

y = x^2 + 5x　を微分すると
y' = 2x + 5

y' = 0 のとき1
0  = 2x + 5 だから
x  = -5/2

そのときのyは
y = -5/2 * -5/2 + 5 * -5/2
  = 25/4 - 25/2
  = -25/4
  
よって最小値は、-25/4となる。
