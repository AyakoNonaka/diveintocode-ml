## 【課題1】データの取得

```python
import pandas as pd
import numpy as np

# irisデータをインポート
from sklearn.datasets import load_iris
iris_dataset = load_iris()

# 説明変数を定義
X = iris_dataset.data
dfX = pd.DataFrame(X)
dfX.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# 目的変数を定義
y = iris_dataset.target
dfy = pd.DataFrame(y)
dfy.columns = ['Species']
```

## 【課題2】データの結合

```python
df = pd.concat([dfX, dfy], axis=1) 
print(df)
```

```python
data = [50, 50, 50]

# ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）

plt.pie(data,
        labels = ['Setosa', 'Versicolor', 'Virginica'],  # ラベル
        startangle=90,  # グラフのスタート位置を90°の位置にしている
        counterclock=False,  # グラフの反時計回りを却下
        autopct='%.2f%%',  # 比率で表示 数字は小数第何位まで表示するかを指定できる
        wedgeprops={'edgecolor': 'black', 'linewidth': 2,},  # 枠線のスタイルの指定
        textprops={'color': 'black', 'weight': 'bold'})  # 文字のスタイルの指定

# グラフの軸を指定できる関数。equalは縦横比を同一にする。
plt.axis('equal')
plt.show()
```
